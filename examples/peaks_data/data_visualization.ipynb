```python
import matplotlib.pyplot as plt
import numpy as np 
import pickle
%matplotlib inline
```

```python
filename = 'classes_20/length_6_points_80.pkl'
```

```python
with open(filename, 'rb') as f:
    data = pickle.load(f)
```

```python
x_train = np.zeros((len(data[0]), 2))
y_train = np.zeros((len(data[0]),))

for i in range(len(data[0])):
    x_train[i, 0] = data[0][i][0]
    x_train[i, 1] = data[0][i][1]
    y_train[i] = data[0][i][2]
    
x_test = np.zeros((len(data[1]), 2))
y_test = np.zeros((len(data[1]),))

for i in range(len(data[1])):
    x_test[i, 0] = data[1][i][0]
    x_test[i, 1] = data[1][i][1]
    y_test[i] = data[1][i][2]
```

```python
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train)
```

```python
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test)
```
